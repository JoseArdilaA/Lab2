```python
import matplotlib.pyplot as plt
import itertools as it

def isortQ(aux):
    v = aux
    steps = 0
    j=0
    x=0
    i=0

    for i in range(len(v)):
        x=v[i]
        j=i-1
        
        while j>-1 and v[j]>x:
            v[j+1]=v[j]
            j=j-1
            steps=steps + 1
        steps=steps + 1
        v[j+1]=x
    return steps

def histoIsort(n):
    k = 1
    total = 0
    iteracion = list(it.permutations(range(n)))
    Tsteps = list(range(len(iteracion)))
    for k in range(len(iteracion)):
        Tsteps[k] = isortQ(list(iteracion[k]))
        total = total + Tsteps[k]
    plt.hist(Tsteps)
    print "Promedio = ", float(total)/len(iteracion)

n = [4,8,12]
```

```python
histoIsort(n[0])
```

```python
histoIsort(n[1])
```
