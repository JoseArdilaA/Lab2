```python
import matplotlib.pyplot as plt
import itertools as it
import math

nperm = 0
cases = list(range(100000))

def isortSteps(aux):
    v = 1*aux
    steps = 0
    j=0
    x=0
    i=0

    for i in range(len(v)):
        x=v[i]
        j=i-1
        while j>-1 and v[j]>x:
            v[j+1]=v[j]
            j=j-1
            steps=steps + 3
        steps=steps+1
        v[j+1]=x
        steps=steps+4
    steps=steps+1
    return steps

def swap (v, i, j):
    t = v[i]
    v[i] = v[j]
    v[j] = t
    
def perm (v, i, n):

    j = 0
    l = 0
    global nperm
    global cases
    
    if (i == n):
        cases[nperm] = isortSteps(v)
        nperm = nperm + 1
    else:
        for j in range(i,n):
            swap (v, i, j);
            perm (v, i+1, n)
            swap (v, i, j);

def histoIsort(n):
    
    k = 0
    total = 0
    v = list(range(n))
    minCase=1000;
    maxCase=0;
    
    for k in range(n):
        v[k]=k+1

    perm(v,0,n);
    Tsteps = list(range(nperm))
    
    for k in range(nperm):
        Tsteps[k] = cases[k]
        total = total + Tsteps[k]
    plt.hist(Tsteps,nperm)
    print "El promedio es: ", float(total)/nperm
    
n = [4,8,12]
```
